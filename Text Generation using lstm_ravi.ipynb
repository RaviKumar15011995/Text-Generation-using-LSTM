```python
from keras.models import Sequential 
from keras.layers import Dense,LSTM,Embedding
from keras.preprocessing.text import Tokenizer # preprocessing
from keras.utils import to_categorical   # target variable
import numpy as np
```

```python
data="rain rain go away come again another day little jhonny wants to play jhonny wants to play again in the rain"
```

```python
tokenizer=Tokenizer()   #it takes list of list
```

```python
tokenizer.fit_on_texts([data])   #it gives word index of each word present in a sentence
```

```python
sequence=tokenizer.texts_to_sequences([data])[0]
```

```python
sequence
```

```python
len(sequence)   #it prints the indexes of each word like 1 -> rain, 2-> go
```

```python
tokens=tokenizer.word_index
```

```python
tokens
```

```python
tokens=len(tokens)+1
tokens
```

Now to prepare dataset for text generation problem<br>
Now the main objective is given a word what is the next word

```python
sequences=[]

for i in range(1,len(sequence)):
    seq=sequence[i-1:i+1]
    print(seq)
    sequences.append(seq)
```

```python
sequences=np.array(sequences)
train_x,train_y = sequences[:,0],sequences[:,1]
```

```python
train_x,train_y
```

```python
train_y=to_categorical(train_y,num_classes=tokens)    #num_classes= no of classes i.e equal to tokens (tokens=tokens+1)
                                                    #because the index starts from zero but in above we started from 1 thats y tokens=tokens+1 happened
                                                    #zero number index is missing
```

```python
len(train_y)
```

Embeddings make it easier to do machine learning on large inputs like sparse vectors representing words.

```python
model=Sequential()
```

```python
model.add(Embedding(tokens,10,input_length=1))   #input_length=1 because we have taken only one sentence
```

```python
model.add(LSTM(32))   #this number can be changed by hyperparameter tuning    ....No of nodes=32
```

```python
model.add(Dense(tokens,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=["accuracy"])   #metrics - to evaluate model
```

```python
model.summary()
```

```python
model.fit(train_x,train_y,epochs=500,verbose=1)    #verbose =1 to check the loss at each epoch/iteration
```

```python
def generate_text(input_text):
    result=input_text
    for j in range(4):
        seq=np.array(tokenizer.texts_to_sequences([input_text]))[0]
        pred_seq=model.predict_classes(seq)
        
        for word,index in tokenizer.word_index.items():
            if index==pred_seq:
                out_word=word
                break
        input_text=out_word
        result+=' '+out_word
    return result
```

```python
generate_text('to')
```

LSTM can generate words given in the tokens only not from the outside of token list
